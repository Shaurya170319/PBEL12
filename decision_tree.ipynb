# Understanding the structure of a Decision Tree

## Trees

- REAL TREE - : It has --> roots --> trunks --> branches --> leaves
- Decision tree: Questions --> more questions --> final answers
- A decision tree grows like a natural tree, only difference is that instead of growing leaves it grows decision


### Parent node:
 - Parent node creates one or more child nodes.
 - Salary(parent node) --> yes/NO -->child node

 ### Child Node
  - Every node created from another node is called Child node

Root Node --> Internal node --> branch --> child node --> leaf node

 - This entire structure is called Decision tree

### Pure Vs Impure nodes

- Basket A: You have [apple,apple,apple,apple]
 - This is pure node.

- Basket B: You have [apple,banana,apple,orange,banana]
 - This is impure node.

- Tree loves pure node.
- The goal is simple convert impure --> less impure --> almost pure --> Completely pure


- **Decision tree grow by increasing purity.**

### Entropy

- Imagine you are playing a guessing game.

- you have a box

-  inside : dog or cat

- Whats inside??

- Dog...........

- You already know the answer.

- There is NO UNCERTAINITY.



- Imagine another box
- It can contain
- Inside: Cat or Dog or Rabbit or Horse or Fish or Bird
- It is difficult to guess because there are many possibilities.

- the UNCERTAINITY has inceased.

- This uncertainity is called **ENTROPY**

- Entropy means "How much confusion or uncertainity exist in the data"

- Less Confusion --> Low Entropy
- More Confusion --> High Entropy
- Decision tree always try to reduce confusion.


Ex - Suppose there are 55 students.

- 55 passed
- 0 failed
- result??
- Entropy = very low

Ex2 - Suppose there are 55 students.
- 27 passed
- 28 failed
- Prediction becomes difficult
- Entropy = high

Ex3 - Suppose there are 55 students.

- 53 passed
- 2 failed
- predict: Passed
- Entropy : Mixed Data --> Low entropy


**Note** 
 - More Mixed data --> High Entropy
 - More Organized Data --> Low Entropy

# Information Gain(How Decision Trees choose the best question)

- Imagine you are a Detective
- A thief has occured in a house.

1. was the door locked or not?
2. was anyone inside or not?
3. fingerprints found??
4. cctv available?
5. ques to ask people?

- Should you ask this question randomly???? No


- because we want to ask **Most** useful info first.

- Decision tree works exactly in the same way

- Which question removes the maximum confusion that reduction in confusion is called "Information Gain"

- How much uncertainity is removed after the splitting the data.

- Ex: suppose : 20 students 
- 10 passed 
- 10 failed
- Entropy --> High
- Now divide the students based on attendance.


- Group A:
- Attendance is > 80%
- 10 passed 
- 2 failed


- Group B:
- Attendance< 80%
- 0 passed
- 8 failed

- Both groups are much cleaner now.
- Prediction becomes easy
- This split has HIGH INFORMATION GAIN.


- Now divide the students based on roll no.
- GROUP a:
- 5 passed
- 5 failed

- Group B:
- 5 passed
- 5 failed

- **Nothing Changed**
- Prediction is still difficult
- Information gain is almost zero

- Good Question ->
- Creates pure groups ->
- Entropy decreases ->
- Information Gain Increases

**Decision tree always selects the question with maximum Information Gain**

### Information gain can be negative??

- No

- In standard Decision tree information gain is non negative because split   should not increase uncertainity overall.

#### Ques 2- Information gain can be ZERO??

- yes
- if splitting on an attribute **does not reduce entropy at all** 

#### Information Gain:
  - The reduction in uncertainity after split

#### Decision tree
 - A sequence of questions leading to a decision

#### Entropy
 - A measure of uncertainity or impurity

#### Best Split
 - The feature with the highest information gain.
