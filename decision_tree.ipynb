# Understanding the structure of a Decision Tree

## Trees

- REAL TREE - : It has --> roots --> trunks --> branches --> leaves
- Decision tree: Questions --> more questions --> final answers
- A decision tree grows like a natural tree, only difference is that instead of growing leaves it grows decision


### Parent node:
 - Parent node creates one or more child nodes.
 - Salary(parent node) --> yes/NO -->child node

 ### Child Node
  - Every node created from another node is called Child node

Root Node --> Internal node --> branch --> child node --> leaf node

 - This entire structure is called Decision tree

### Pure Vs Impure nodes

- Basket A: You have [apple,apple,apple,apple]
 - This is pure node.

- Basket B: You have [apple,banana,apple,orange,banana]
 - This is impure node.

- Tree loves pure node.
- The goal is simple convert impure --> less impure --> almost pure --> Completely pure


- **Decision tree grow by increasing purity.**

### Entropy

- Imagine you are playing a guessing game.

- you have a box

-  inside : dog or cat

- Whats inside??

- Dog...........

- There is NO UNCERTAINITY.



- Imagine another box

**Note**

# Gini Index (The Fast Decision maker)

- Imagine there are two judges in a competition.


- Judge A:
- Think very carefully.
- Checks every rule.
- Overall calculate everything.
- Finally tells who is winner.

 

Judge B:
- Looks quickly
- makes fast decision.
- stills gives the almost same winner.

- Who finishes first??
- Judge B

- This is exactly the difference between **Entropy** and **Gini Index**

- Both measure Impurity
- but Gini Index is **computationally faster**.
- That is why the **CART algorithm uses GINI INDEX**


#### Gini Index???
- Gini Index measures how much impure  or mixed dataset is.
- More mixed data --------->>> Higher Gini
- More pure data ->>>>>> Lower Gini
- **Decision tree always try to reduce GINI INDEX**

- Entropy ask *"How much information is missing?"*
- Gini ask *"How likely am I to classify a incorrectly?"*
- Both reached the almost same decision
- But gini requires fewer calculation
- Therefore, Gini is faster.

- Baskets:


- **Basket A**: apple,apple,apple,apple
- only apple
- Gini = 0
- because Gini works towards more purity.
- Output Data is completely pure. 

**Basket B**:  Apple,apple,banana,apple,banana
- Half apples
- Half bananas
- Now mistake can be possible
- Gini is Higher

**Basket C**: Apple, banana ,grapes, watermelon, oranges
- everything mixed
- prediction becomes difficult
- Highest impurity

- binary classification:
- gini =1 -(p1^2 +p2^2)
- p1 = probability of class 1
- p2 = probability of class 2

- for multiple classes
- gini = 1 - $\sum$(pi^2)

- dataset: 10 student
- pass - 10
- fail - 0

- probability of pass = 10/10=1
- probability of fail = 0/10 =0

- gini = 1 - (1+0)
- gini =0
  - therefore it is perfectly pure dataset.

- dataset 2
- fail =5
- pass= 5
- probability of pass = 0.5
- probability of fail = 0.5
- gini = 1- (0.25+0.25)
- gini = 0.5
- Max gini for binary classification

- Customer 100
- purchase= 80
- not purchased = 20
- prob of purchase =0.8
- prob of not purchased = 0.2


gini = 1- (0.64 + 0.04)= 0.32

- dataset is fairly pure

```python
def gini(p,n):
    total = p + n
    if total == 0:
        return 0
    return 1 - (p/total)**2 - (n/total)**2
```

```python
print("Pure Gini", gini(10, 0))
print("Balanced Gini", gini(5,5))
print("80-20 Gini", round(gini(10, 5),3))
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt# pyplot for plotting graphs.
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split # shuffle + train and split the data
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,precision_score, recall_score, f1_score
```

```python
cancer = load_breast_cancer() 
```

```python
df= pd.DataFrame(cancer.data, columns = cancer.feature_names)
df['Target'] = cancer.target # Assigns the target variable to the DataFrame
```

```python
df.head(10)
```

```python
print("shape :" , df.shape)
```

```python
print(df.info())
```

```python
print(df['Target'].value_counts()) # Check the distribution of the target variable
```

```python
print(df.isnull().sum()) # Check for missing values
```

```python
x= df.drop('Target',axis=1)
y=df['Target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,y, test_size =0.25,random_state = 42
)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

### Creation of Decision tree model

```python
model = DecisionTreeClassifier(
    criterion='gini', # or 'entropy'
    random_state =42
)
```

```python
model
```

#### Training model -:

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
comparison = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred
})
```

```python
comparison.head(10)
```

```python
accuracy = accuracy_score(y_test,y_pred)
```

```python
round(accuracy,2)
```

```python
precision =precision_score(y_test,y_pred)
round(precision,2)
```

```python
recall = recall_score(y_test,y_pred)
round(recall,2)
```

```python
f1 = f1_score(y_test,y_pred)
round(f1,2)
```

```python
confusion_mat = confusion_matrix(y_test,y_pred)
(confusion_mat)
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(confusion_mat, annot=True, cmap='Blues', xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['Actual Negative', 'Actual Positive'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
