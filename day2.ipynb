```python
#data types
#list
my_list = [1, 2, 3, 4, 5]
print(my_list)  # Output: [1, 2, 3, 4, 5]

# Methods 
print(my_list)
my_list.append(6)  # Adds an element to the end of the list
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]


my_list.insert(2, 10)  # Inserts 10 at index 2
print(my_list)  # Output: [1, 2, 10, 3, 4, 5, 6]


my_list.remove(3)  # Removes the first occurrence of 3
print(my_list)  # Output: [1, 2, 10, 4, 5, 6]


my_list.pop()  # Removes and returns the last element
print(my_list)  # Output: [1, 2, 10, 4, 5]


# properties
#ordered and mutable
my_list[2]= 20  # Modifies the element at index 2
print(my_list)  # Output: [1, 2, 20, 4, 5]


# heterogeneous & duplicates allowed
zen=[1,2,3,4,5,1,2,3,"ai",True,False]
print(zen)  # Output: [1, 2, 3, 4, 5, 1, 2, 3, 'ai', True, False]


```

```python
#var 1=45
#function

class Car:
    #attributes  color,model,year
    def __init__(self, color, model, year):
        self.color = color
        self.model = model
        self.year = year
    #methods start stop play
    def start(self):
        print("The car is starting.")

    def stop(self):
        print("The car is stopping.")

    def play(self, song):
        print(f"Playing {song}.")
    
bmw = Car("Red", "BMW X5", 2022)
```

```python
bmw.start()  # Output: The car is starting.
bmw.play("Shape of You")  # Output: Playing Shape of You.
bmw.stop()  # Output: The car is stopping.
```

```python
# Create a class box give some attribute and box placing method stack push and pop



class Box:
    def __init__(self, color, size):
        self.color = color
        self.size = size
        self.items = []
    def stack(self, item):
        self.items.append(item)
        print(f"Stacked {item} in the box.")
    def push(self):
        if self.items:
            item = self.items.pop()
            print(f"Popped {item} from the box.")
        else:
            print("The box is empty.")
    def place(self):
        print(f"The box is placed. Color: {self.color}, Size: {self.size}, Items: {self.items}")
box1 = Box("Blue", "Large")
```

```python
box1.stack("Book")  # Output: Stacked Book in the box.
box1.stack("Laptop")  # Output: Stacked Laptop in the box.  
box1.place()  # Output: The box is placed. Color: Blue, Size: Large, Items: ['Book', 'Laptop']
box1.push()  # Output: Popped Laptop from the box.
```

```python
while True:
    user_input = input("Enter a number (or 'exit' to quit): ")
    if user_input.lower() == 'exit':
        break
    try:
        number = int(user_input)
        print(f"You entered: {number}")
    except ValueError:
        print("Please enter a valid number.")
```
