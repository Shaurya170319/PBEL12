```python
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.image as img
```

```python
(x_train,Y_train),(x_test,Y_test)=keras.datasets.mnist.load_data()
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train_flattened = x_train.reshape(len(x_train),28*28)
x_test_flattened = x_test.reshape(len(x_test),28*28)
```

```python
x_train_flattened.shape
```

```python
x_train_flattened[0]
```

```python
model= keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train_flattened,Y_train,epochs=8)
```

```python
model.evaluate(x_test_flattened,Y_test)
```

```python
Y_test[0]
```

```python
y_predicted= model.predict(x_test_flattened)
y_predicted[0]
```

```python
plt.matshow(x_test[0])
```

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_labels= [np.argmax(i) for i in y_predicted]
```

```python
y_predicted_labels[:5]
```

```python
cm= tf.math.confusion_matrix(labels= Y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

USING HIDDEN LAYER

```python
model1 = keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model1.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics= ['accuracy']
    )

model1.fit(x_train_flattened,Y_train,epochs=10)
```

```python
model1.evaluate(x_test_flattened,Y_test)
```

```python
y_predicted1 = model1.predict(x_test_flattened)
y_predicted_labels1 = [np.argmax(i) for i in y_predicted1]
cm1 = tf.math.confusion_matrix(labels=Y_test,predictions=y_predicted_labels1)


plt.figure(figsize=(10,7))
sn.heatmap(cm1,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

Using Flatten layers so that we don't have to call .reshape on input dataset

```python
model12 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model12.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics= ['accuracy']
    )

model12.fit(x_train,Y_train,epochs=5)
```

```python
model12.evaluate(x_test,Y_test)
```
