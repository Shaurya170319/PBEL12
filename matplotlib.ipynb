```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# univariate, bivariate, multivariate analysis
# univariate analysis - analysis of a single variable/column such as mean, median, mode, variance, standard deviation,age,hist,categorical etc.
# bivariate analysis - analysis of two variables/columns such as correlation, scatter plot, line plot, bar plot, box plot etc.
# multivariate analysis - analysis of multiple variables/columns such as heatmap, pair plot, etc.
```

```python
import pandas as pd
tips = pd.read_csv('Data/tipsdataset.csv')
```

```python
tips
```

```python
#univariate analysis
plt.plot(tips['total_bill'])# line plot used to show the relationship between two variables
plt.xlabel('total bill')# x-axis label
plt.show()
```

```python
plt.scatter(tips['size'], tips['total_bill'],s=tips['total_bill']) # scatter plot used to show the relationship between two variables
plt.title("scatter") # title of the plot
plt.show()
```

```python
 # add categorical column count plot
plt.pie(tips['gender'].value_counts(),autopct='%1.1f%%') # pie chart used to show the distribution of a single categorical variable
plt.show()
```

```python
plt.hist(tips['total_bill'], bins=10) #univariate analysis - histogram used to show the distribution of a single variable
# [1,20,30,40,50]divide data into 4 parts ,1-20,21-30,31-40,41-50
plt.show()
```

```python
sns.relplot(data=tips, x="tip", y="size",col='gender',hue='smoker',size='total_bill')
plt.show()
```

```python
# Create a visualization
sns.relplot(
data=tips,
x="total_bill", y="tip", col="time",
hue="smoker", style="smoker", size="size",
)
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)#kde= kernel density estimation, a way to estimate the probability density function of a random variable
# uses of this= to visualize the distribution of a single continuous variable and estimate its probability density function
#this is univariate analysis
sns.displot(data=tips, x="tip", col="time", kde=True)
sns.displot(data=tips, x="gender", col="time", kde=True)
plt.show()
```
