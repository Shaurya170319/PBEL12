# Types Of Classification

# 1.Binary Classification 

When the target value has EXACTLY two possible outcomes. It comes under binary(0,1)


Exmaples
1. Email Spam detection
2. Medical diagnosis(Diseased or healthy)
3. Application : Logistic Regression,SVM ,and Naive Bayes

# 2. Multi Class Classification

When Dataset need to be categorized into  more than two distinct classes.

Multiple categories available each data point can still only be assigned to a Sinle or final category.


- Exmaple 1 : We have 
    - Animal : Dog,Cat ,Horse,Lion
    - Still it can only assigned to a single outcomes : Cat orDog or Horse or Lion


- Example 2 : we have 
    - A,B,C,D
    - Multi - Class

- Algorithm:
   - Decision Tree,Random Forest, and Naive Bayes

# Confusion Matrix
 It is a small table that has actual labels with predicted labels.

1. True Positive: Predicted positive,actually positive.
2. True Negative: Predicted negative, actually Negative.
3. False Positive: Predicted Positive,actually negative.
4. False Negative: predicts negative , actually positive.

- Examples : Emai is spam
    - Actual spam, predicted spam = TP
    - Actual not spam,predicted spam = FP
    - Actual spam,predicted not spam= FN
    - Actual not spam,Predicted not spam= TN

# Accuracy
Accuracy tells us how much model is correct overall.

- Example: if 100 emails,the model gets 90 right, accuracy 90%



- When it will Mislead:
   - Accuray can look high even when the model is poor. on the important class.
   - Example: If only 1 out of 100 email is spam , then the model that always say not spam not spam ->>>>>> Accuracy 99%
   - Useless for finding spam

# Precision:
Meaning-:
 When the model predicts POSITIVE, right??
 - precision = TP/TP+FP

- if spam filter marks 10 emails as spam and 8 really are spam.
- precision= 8/10 = 80%

# Recall


- out of all real Positive cases, How may did the model finds
- precision = TP/TP+FN

- Example:
  - there are only 12 spam emails and model catches 9 of them
  - 9/12 = 75%
